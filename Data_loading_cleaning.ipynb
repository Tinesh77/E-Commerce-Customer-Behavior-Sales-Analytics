```python
import numpy as np
import pandas as pd

df = pd.read_csv("online_retail.csv")
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['Revenue'] = df['Quantity']*df['UnitPrice']
```

```python
print(df)
type(df['InvoiceDate'])
```

```python
df = df.dropna(subset=['CustomerID'])
print(df)
```

```python
df = df[df['Quantity'] > 0]
print(df)
```

```python
df.to_csv("cleaned_data.csv",index=False)
```
